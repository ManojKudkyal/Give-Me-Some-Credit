### See there is a lot of difference.(unbalanced)
### If we give this dataset to our model, model will be biased.
### i.e. it will provide good results for train data, but will be less accurate on test data or when it sees the new inputs.

# **So what will we do is we take a fraction of the records of the SeriousDlqin2yrs = 0 from dataset. After that the dataset will be balanced.**

```python
train_serious_0 = train_data[train_data['SeriousDlqin2yrs']==0].sample(frac=0.07)
```

```python
train_serious_1 = train_data[train_data['SeriousDlqin2yrs']==1].copy()
```

```python
balanced_train_data = pd.concat([train_serious_0,train_serious_1],axis=0)
balanced_train_data.head()
```

```python
sns.countplot(x='SeriousDlqin2yrs',data=balanced_train_data)
```

# **Now after all the cleaning data and EDA, let's train our model.**
# **We will build model for both unbalanced and balanced datasert**
---



```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_curve, auc, log_loss
from sklearn.linear_model import LogisticRegression
```

Droping unwanted columns.

```python
cols = list(final_train.drop(['Unnamed: 0','SeriousDlqin2yrs'],axis=1))
cols
```

```python
X = final_train[cols]
y = final_train['SeriousDlqin2yrs']
```

```python
X1 = balanced_train_data[cols]
y1 = balanced_train_data['SeriousDlqin2yrs']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=2)
```

```python
LogisticModel = LogisticRegression()
```

```python
LogisticModel1 = LogisticRegression()
```

```python
LogisticModel.fit(X_train, y_train)
```

```python
y_pred = LogisticModel.predict(X_test)
```

```python
y_pred_proba = LogisticModel.predict_proba(X_test)[:, 1]
```

```python
[fpr, tpr, thr] = roc_curve(y_test, y_pred_proba)

print('Train/Test split results:')
print(LogisticModel.__class__.__name__+" accuracy is %2.3f" % accuracy_score(y_test, y_pred))
print(LogisticModel.__class__.__name__+" log_loss is %2.3f" % log_loss(y_test, y_pred_proba))
print(LogisticModel.__class__.__name__+" auc is %2.3f" % auc(fpr, tpr))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

# **Acccuracy score is quite good.**
# **But there is a bias in the model. As F1-score for 0 is 0.97 and for 1 it is only 0.02**
