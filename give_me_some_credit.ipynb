```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_curve, auc, log_loss
from sklearn.linear_model import LogisticRegression
```

Droping unwanted columns.

```python
cols = list(final_train.drop(['Unnamed: 0','SeriousDlqin2yrs'],axis=1))
cols
```

```python
X = final_train[cols]
y = final_train['SeriousDlqin2yrs']
```

```python
X_balaced = balanced_train_data[cols]
y_balaced = balanced_train_data['SeriousDlqin2yrs']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train_balaced, X_test_balaced, y_train_balaced, y_test_balaced = train_test_split(X_balaced, y_balaced, test_size=0.3, random_state=42)
```

```python
LogisticModelUnbalance = LogisticRegression()
```

```python
LogisticModelUnbalance.fit(X_train, y_train)
```

```python
y_pred_unbalanced = LogisticModelUnbalance.predict(X_test)
```

```python
y_pred_proba = LogisticModelUnbalance.predict_proba(X_test)[:, 1]
```

```python
[fpr, tpr, thr] = roc_curve(y_test, y_pred_proba)

print('Train/Test split results:')
print(LogisticModelUnbalance.__class__.__name__+" accuracy is %2.3f" % accuracy_score(y_test, y_pred_unbalanced))
print(LogisticModelUnbalance.__class__.__name__+" log_loss is %2.3f" % log_loss(y_test, y_pred_proba))
print(LogisticModelUnbalance.__class__.__name__+" auc is %2.3f" % auc(fpr, tpr))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_unbalanced))
```
